1. What is the difference between a function and a method in Python?
-> **Function**: A function is a block of reusable code that is designed to perform a specific task.
It can be defined using the `def` keyword and can be called independently of any object.
Functions can accept arguments and return values. - Example: `def add(x, y): return x + y` - **Method**: A method is similar to a function but it is associated with an object. It is a function
that belongs to an object or class and is called using the dot notation. Methods usually operate
on the data contained within the object. - Example: `my_list.append(5)` (Here, `append()` is a method of a list object.)

6. Explain the concept of generators in Python and how they are defined?
->  What is a Generator in Python?
A generator is a special type of function that yields values one at a time instead of returning them all at once. It allows lazy evaluation, meaning it produces values on the fly and does not store them in memory.
Generators are memory efficient, especially for large datasets or infinite sequences.
How to Define a Generator
Generators are defined using def like a normal function, but instead of return, they use the yield keyword.

7. What are the advantages of using generators over regular functions?
-> Generators have several advantages, especially when dealing with large data, streams, or infinite sequences.
 Advantages of Generators
1. Memory Efficiency
Generators do not store the entire sequence in memory.
They yield items one by one, which saves RAM.
2. Lazy Evaluation
Values are computed only when needed, which can boost performance.
Regular functions return everything at once, but generators wait until you ask for the next value.
3. Faster Performance with Large Data
Since data is yielded as needed, no time is wasted building huge lists upfront.
4. Pipeline-Friendly
Generators work well in data pipelines, where each step processes data and passes it along.
5. Cleaner Code for Iterators
Easier to write custom iterators using yield instead of creating a class with __iter__() and __next__().

4. Write a Python function that checks if a given number is prime or not from 1 to 200?

````python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
    for num in range(1, 201):
        if is_prime(num):
          print(f"{num} is a prime number")
        else:
          print(f"{num} is not a prime number")
````

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms?


````python
class Fibonacci:
    def __init__(self, max_value):
        self.max = max_value
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.a > self.max:
            raise StopIteration
        fib = self.a
        self.a, self.b = self.b, self.a + self.b
        return fib
fib_seq = Fibonacci(100)
for num in fib_seq:
    print(num, end=" ")


````

6. Write a generator function in Python that yields the powers of 2 up to a given exponent?

````python
def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i
for val in powers_of_two(5):
    print(val, end=" ")
````

7. Implement a generator function that reads a file line by line and yields each line as a string?

````python
def file_reader(filename):
    """
    Reads a file line by line and yields each line as a string.
    """
    try:
        with open(filename, 'r') as file:
            for line in file:
                yield line.strip()
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
        yield None
````

10. Create a Python program that uses `filter()` to remove all the vowels from a given string?

````python
def remove_vowels(text):
    vowels = "aeiouAEIOU"
    return "".join(filter(lambda ch: ch not in vowels, text))
input_str = "Hello World"
result = remove_vowels(input_str)
print("String without vowels:", result)
````

Write a Python program using lambda and map.

````python
# Sample order data
orders = [
    ("1.34587", "learning python, mark lutz", 4),
    ("2.98762", "programming python, mark lutz", 5),
    ("3.77226", "headd first python, paul barry", 3),
    ("4.88112", "einfuhrung in python3, brend klein", 3)
]

# Assume price per book is €20
price_per_book = 20

# Function to calculate total order price with conditions
calculate_price = lambda order: (
    order[0],  # Order number
    (order[2] * price_per_book + 10) if (order[2] * price_per_book < 100) else (order[2] * price_per_book)
)

# Using map() to apply function to all orders
order_totals = list(map(calculate_price, orders))

# Output results
print(order_totals)
````
